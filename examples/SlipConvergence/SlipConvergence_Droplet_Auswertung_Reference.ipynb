```
public static Dictionary<string, Plot2Ddata> ExtractConvergenceH(IEnumerable<ISessionInfo> sessions, int k, double theta, double ls, double q, double sigma, bool recoil, bool convection, bool quadratic){

    Dictionary<string, Plot2Ddata> ret = new Dictionary<string, Plot2Ddata>();
    
    var sess = sessions.Where(s => 
        Convert.ToDouble(s.KeysAndQueries["id:wallfriction"]) == 0.0 &&
        Convert.ToDouble(s.KeysAndQueries["id:theta"]) == Math.Round(theta, 4) &&
        Convert.ToDouble(s.KeysAndQueries["id:interfacesliplength"]) == ls &&
        Convert.ToDouble(s.KeysAndQueries["id:heatflux"]) == q &&
        Convert.ToDouble(s.KeysAndQueries["id:sigma"]) == sigma &&
        Convert.ToBoolean(s.KeysAndQueries["id:recoil"]) == recoil &&
        Convert.ToBoolean(s.KeysAndQueries["id:convection"]) == convection && 
        Convert.ToBoolean(s.KeysAndQueries["id:quadraticls"]) == quadratic && 
        (k > 0 ? (Convert.ToInt32(s.KeysAndQueries["id:degree"]) == k) : true)
    ).ToArray();
    
    ITimestepInfo[] timesteps = sess.Select(s => s.Timesteps.Last()).ToArray();
    // string[] fields = timesteps.First().Fields.Select(f => f.Identification).ToArray();
    string[] fields = new string[] {"Pressure", "VelocityX", "VelocityY"};
    
    foreach(var field in fields){
        Plot2Ddata errDat = sess.ToEstimatedGridConvergenceData(field, normType: NormType.L2_embedded); 
        ret.Add(field, errDat);
    }
    
    Console.WriteLine(sess.Count());
    
    return ret;
}
```

### StaticDropletOnWall Sessions bis mesh5, Kontaktwinkel 90°

```
// Testcase setup
double[] thetaS            = (new double[] {90}).Select(x => x/180.0 * Math.PI).ToArray();
double[] lsS               = new double[] {0.0};
double[] qS                = new double[] {0.5};
double[] sigmaS                = new double[] {0.1};
bool[] recoilS = new bool[] {false}; // recoil pressure
bool[] convectionS = new bool[] {false}; // convection
bool[] quadraticS = new bool[] {true, false}; // convection

//BERECHNUNG
Dictionary<(double, double, double, double, bool, bool, bool), Dictionary<string, Plot2Ddata>> ConvergenceData = new Dictionary<(double theta, double ls, double q, double sigma, bool recoil, bool convection, bool quadratic), Dictionary<string, Plot2Ddata>>();

foreach(double theta in thetaS){
    foreach(double ls in lsS){
        foreach(double q in qS){
            foreach(double sigma in sigmaS){
                foreach(bool recoil in recoilS){
                    foreach(bool convection in convectionS){
                        foreach(bool quadratic in quadraticS){
                            ConvergenceData.Add((theta: theta, ls: ls, q: q, sigma: sigma, recoil: recoil, convection: convection, quadratic: quadratic), ExtractConvergenceH(sessions, 0, theta, ls, q, sigma, recoil, convection, quadratic));
                        }
                    }
                }
            }
        }
    }
}
```

```
foreach(var data in ConvergenceData){
    
    Console.WriteLine(data.Key.ToString());    
    
    var dats = data.Value;
    
    foreach(string field in dats.Keys){
        Console.WriteLine(field);
        var regression = dats[field].Regression().OrderBy(k => k.Key);
        foreach(var kvp in regression){
            Console.WriteLine($"{kvp.Key} : {kvp.Value}");
        }
    }
    
    
    Console.WriteLine();
}
```
