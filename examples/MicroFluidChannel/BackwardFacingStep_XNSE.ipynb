## Problem settings

```
double density = 1000.0;
double viscosity = 1.0e-3;
double freestreamVelocity = 1.0;
double stepHeight = 100.0e-6;

double Re = density * freestreamVelocity * stepHeight / viscosity;
Re
```

```
Formula InletVelocity = new Formula("(X) => 1.0", false);
```

## Grid Creation

Construct a grid consisting of three parts:
1. Inflow
2. OuterFlow
3. StepFlow

```
int upStreamFactor = 4;
double upStreamLength = upStreamFactor * stepHeight;
double upStreamHeight = 8.0 * stepHeight;
double outflowLength = 20.0 * stepHeight;
double TotalHeight = stepHeight + upStreamHeight;
double TotalLength = upStreamLength + outflowLength;
```

```
int upStreamLengthResolution = 4;
int freeStreamHeightResolution = 4;
int freeStreamLengthResolution = 8;
int stepResolution = 8;
```

```
public static double[] GeometricSpacingGrid(double min, double max, int numberOfElements, double growthRatio) {
    double scalingFactor = 0;
    for (int i = 0; i < numberOfElements; i++) {
        scalingFactor += Math.Pow(growthRatio, i);
    }
    double initialCellSize = (max - min) / scalingFactor;
    var nodes = new List<double> { min };
    for (int i = 0; i < numberOfElements; i++) {
        nodes.Add(nodes.Last() + initialCellSize * Math.Pow(growthRatio, i));
    }
    return nodes.ToArray();
}
```

```
double[] xNodes1 = GenericBlas.Linspace(0, upStreamLength, upStreamFactor*upStreamLengthResolution + 1);
double[] yNodes1 = GenericBlas.Linspace(stepHeight, TotalHeight, 8*freeStreamHeightResolution + 1);

var grd1 = Grid2D.Cartesian2DGrid(xNodes1, yNodes1);
```

```
double[] xNodes2 = GenericBlas.Linspace(upStreamLength, TotalLength, 20*freeStreamLengthResolution + 1);
double[] yNodes2 = yNodes1; //GenericBlas.Linspace(stepHeight, TotalHeight, 8*freeStreamResolution + 1);

var grd2 = Grid2D.Cartesian2DGrid(xNodes2, yNodes2);
```

```
double[] xNodes3 = xNodes2; //GenericBlas.Linspace(upStreamLength, TotalLength, 20*freeStreamResolution + 1);
double[] yNodes3 = GenericBlas.Linspace(0, stepHeight, stepResolution + 1);

var grd3 = Grid2D.Cartesian2DGrid(xNodes3, yNodes3);
```

```
var gridMerged = GridCommons.MergeLogically(new GridCommons[] {grd1, grd2, grd3});
var grd = GridCommons.Seal(gridMerged, 4);
```

```
grd.NumberOfCells
```

```
grd.DefineEdgeTags(delegate(Vector X) {
    string ret = null;
    if (X.x.Abs() <= 1e-8)
        ret = IncompressibleBcType.Velocity_Inlet.ToString();
    if ((X.y - TotalHeight).Abs() <= 1e-8)
        ret = IncompressibleBcType.FreeSlip.ToString();
    if ((X.x - TotalLength).Abs() <= 1e-8)
        ret = IncompressibleBcType.Pressure_Outlet.ToString();
    if ((X.x < upStreamLength) && (X.y - stepHeight).Abs() <= 1e-8)
        ret = IncompressibleBcType.Wall.ToString();
    if ((X.x - upStreamLength).Abs() <= 1e-8 && (X.y < stepHeight))
        ret = IncompressibleBcType.Wall.ToString();
    if ((X.x > upStreamLength) && X.y.Abs() <= 1e-8)
        ret = IncompressibleBcType.Wall.ToString();
    return ret;
}); 
```

## Set up Controls 

```
int k = 2;
```

```
List<XNSE_Control> Controls = new List<XNSE_Control>();
Controls.Clear();
```

```
XNSE_Control C = new XNSE_Control();

C.SetGrid(grd);
C.SetDGdegree(k);

C.PhysicalParameters.rho_A = density;
C.PhysicalParameters.mu_A = viscosity;
C.PhysicalParameters.IncludeConvection = true;

C.AddBoundaryValue(IncompressibleBcType.Velocity_Inlet.ToString(), "VelocityX#A", InletVelocity);

C.TimesteppingMode = AppControl._TimesteppingMode.Steady;
C.Option_LevelSetEvolution = LevelSetEvolution.None;

C.SessionName = "BFS_SetupTest4";

Controls.Add(C);
```

## Launch Jobs

```
Controls.Select(C => C.SessionName)
```

```
foreach(var ctrl in Controls) {
    var oneJob              = ctrl.CreateJob();

    oneJob.NumberOfMPIProcs = 1;

    int numThreads = 4;
    oneJob.NumberOfThreads = numThreads;
    IDictionary<string, string> args = oneJob.EnvironmentVars;
    args["BOSSS_ARG_7"] = numThreads.ToString();
    args.Remove("OMP_NUM_THREADS");
    oneJob.MySetCommandLineArguments(args.Values.ToArray());

    oneJob.Activate(myBatch); 
}
```
